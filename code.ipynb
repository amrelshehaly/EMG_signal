```python
import numpy as np
from numpy import loadtxt
from matplotlib.pyplot import plot, show, title
from time import gmtime, strftime
from scipy.signal import butter, lfilter , find_peaks , filtfilt
import matplotlib.pyplot as plt
from scipy import stats
from scipy import signal
import math



def readfile():
                
    emg = loadtxt("Data.txt",)
#     plotgraph(emg[30000:35000])    
    return emg[30000:35000]

def plotgraph(data,label):
    title("Original signal")
    plot(data)
    plt.savefig(label)
    show()
    
def rectified(data):
    ab = []
    for x in data:
        ab.append(abs(x))
    return ab

def threshold(data):
#     print(data)
    return np.std(data[0:500],dtype = np.float32) * 3

def moving_avg_window(data,N):
    return np.convolve(data, np.ones((N,))/N, mode='valid')

def eliminating_False_detection(data , threshold , initial_data):
    output = []
    draft = []
    output_val = []
    T = 20;
    flag = False
    
    prev_count = 0;
    counter = 0
    while(counter < len(data)):
        
        draft = []
        for y in range(counter , len(data)):
            
            if data[y] >= threshold:
                
                draft.append(y)
            else:
                if len(draft) >= T:
                    flag = True
                counter = y+1
                break
                
        if flag :
            
            
            max_indx = 0
            max_value = 0
            for z in range(len(draft)):
                if initial_data[draft[z]] >= max_value:
                    max_indx = draft[z]
                    max_value = initial_data[draft[z]]

            output.append(max_indx)
            flag = False
        
        if prev_count != counter:
            prev_count = counter
        else:
            counter+=1
            
    return output
        
def getting_T_points(data , indxs):
    output = []
    for x in range(len(indxs)):
        draft = []
        i = indxs[x]-10
        j = indxs[x]+10

        for y in range(i,j):
            draft.append(data[y])
        output.append(draft)

    return output

def diff_threshold(data):
    templates = []
    Diff_thres = 12.65**5
    templates.append(data[0])
    
    class_arr = []
    
    for x in range(0,len(data)):
        class_arr.append(-1) 
    
    for x in range(len(data)):
        minmum = 100**12
        temp_no = -1
        for y in range(len(templates)):

            if calc_D(data[x],templates[y]) < minmum :
                minmum = calc_D(data[x],templates[y])
                temp_no = y


        if minmum < Diff_thres :
            class_arr[x] = temp_no
            avg = average(data[x], templates[temp_no])
            templates[temp_no]=avg
        else:
            templates.append(data[x])
            class_arr[x] = len(templates) - 1

    return class_arr , templates
        

def placing_color(axis,templates,indxs):
    plt.figure(figsize=(12,6))
    
    reds =[]
    greens = []
    for x in range(len(templates)):
        if templates[x] == 0 :
            reds.append(indxs[x])
        else:
            greens.append(indxs[x])
    
    print(len(greens))
    print(len(reds))
    plt.plot(axis)
    plt.plot(reds, axis[reds], "*",color="red")
    plt.plot(greens, axis[greens], "*",color="green")
    plt.savefig('DetectedMUAP.jpg')
    plt.show()


def average(a,b):
    result = []
    for x in range(len(a)):
        suming = a[x] + b[x]
        divs = suming  / 2
        result.append(divs)
    return result
        
def calc_D (a,b):
    D = 0
    difference = []
    
    zip_object = zip(a, b)
    for x, y in zip_object:
        difference.append((x-y)**2)
        
    for z in difference:
        D+=z
    return D


def plot_data(axis, dis, lab, stars_plot):
    plt.figure(figsize=(12,6))
    np.diff(stars_plot)
    plt.plot(axis)
    plt.title(lab)
    plt.plot(stars_plot, axis[stars_plot], "*")
    plt.show()


if __name__=="__main__":
    a = readfile()
    b = rectified(a)
    d = moving_avg_window(b,20)
    c = threshold(d)
    
    e = eliminating_False_detection(d , c, a)
    u = getting_T_points(a, e)

    plot_data(a,c,"putting stars", e)
    
    data_temp, fif = diff_threshold(u)
    print(data_temp)
    
    placing_color(a,data_temp,e)
    
    
    plotgraph(fif[0],'template_1.jpg')
    plotgraph(fif[1],'template_2.jpg')
```
